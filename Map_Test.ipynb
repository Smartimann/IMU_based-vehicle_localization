```python
def get_bounding_box_as_points(position : np.ndarray, extent: np.ndarray): 
    box = []
    for y in np.arange(position[1], position[1] + extent[1], 0.1): 
        for x in np.arange(position[0], position[0] + extent[0]): 
            box.append(np.array([x,y]))
    return np.array(box)
```
