```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

%matplotlib inline
```

```python
sim_data = pd.read_csv("../data/Simulation_artificial_noise.csv")
```

```python
sim_data.head()
```

```python
std = np.array([.2,.2,.2])
sim_data['acc_x_noise'] = sim_data['accelerometer_x'] * std[0]
sim_data['acc_y_noise'] = sim_data['accelerometer_y'] * std[0]
sim_data['acc_z_noise'] = sim_data['accelerometer_z'] * std[0]
```

```python
sim_data.head()
```

```python
sim_data.drop(['acc_y', 'acc_x','acc_z'], axis=1)
```

```python
acc_norms = []
for i in range(len(sim_data['accelerometer_x'])): 
    acc_norms.append(np.linalg.norm(np.array(sim_data[sim_data['accelerometer_x'][i], sim_data['accelerometer_y'][i]])))

sim_data['acceleration'] = acc_norms
```

```python
sim_data.to_csv("../data/noisedSimData.csv")
```
